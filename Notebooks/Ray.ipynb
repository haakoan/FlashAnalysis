```python
import yt
import numpy as np
```

```python
ck_f = 's20_0.75_hdf5_plt_cnt_1595'
ds = yt.load(ck_f)
```

```python
#Cast a ray from P1 to P2, in this case (0,0,0) and (1e9,0,0).
ray = ds.ray((0,0,0),(1e9,0,0))

#You can now get the data from the ray as you would a dict
# ie entropy = ray['entr']
```

```python
import matplotlib                                                                                                                           
import matplotlib.pyplot as plt

f,ax = plt.subplots(figsize=(7,5))
plt.plot(ray["r"],ray['dens'],lw=2,label="dens")
ax.set_xlim([0,150e5])
ax.set_xlabel("r [cm]")
ax.legend(loc=0)
```
