```python
import yt
import numpy as np
```

```python
#Adjust to your file
ck_f = 's20_0.75_hdf5_plt_cnt_1595'
ds = yt.load(ck_f) #Loads the data
```

```python
#Simple plot of a 2D simulation

#Set window size, this setup works well for 2D
ywindow = 2e7
xwindow = ywindow/1.8

#Decide what to plot
var = 'dens'
slc = yt.SlicePlot(ds,'theta',var,center=[xwindow/2, 0.0,np.pi],
                   width=((xwindow,'cm'), (ywindow, 'cm')))


slc.set_cmap(var, "RdBu") #here we plot the field set in var
#Set the density range on the colorbar
slc.set_zlim(("dens"), zmin=(1e8, "g/cm**3"), zmax=(1e12, "g/cm**3")) 
slc.set_figure_size(7)

slc.show() #save with .save()
```
